# **Salaries of Data Scientists in 2020-2023**

## **1. Data Overview**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('Data Science Salary 2021 to 2023.csv')
df
```

```python
df.info()
```

## **2. Impact of Employment Type and Job Title to the Salary**

**2.1. Average, Highest and Lowest Salaries**

```python
vid_atl = round(df['salary_in_usd'].mean(),2)
max_atl = round(df['salary_in_usd'].max(),2)
min_atl = round(df['salary_in_usd'].min(),2)
print(f'Average Salary - {vid_atl} USD')
print(f'Highest Salary - {max_atl} USD')
print(f'Lowest Salary - {min_atl} USD')
```

**2.2. Average Salary According to Different Employment Types**

```python
et_vid_atl = df.groupby('employment_type').mean()['salary_in_usd'].round(2).sort_values(ascending = False)
et_vid_atl
```

```python
et_vid_atl.plot(kind = 'bar', color = 'darkgreen', width = 0.8)

plt.title('Average Salary According to Different Employment Types in USD', fontdict={'fontweight':'bold', 'fontsize': 10})
plt.xlabel(' ')
plt.show()
```

```python
ft_df_sorted = df[df['employment_type'] == 'FT'].sort_values(by='salary_in_usd', ascending=False)
ft_top = ft_df_sorted[:10]
ft_vid = round(ft_df_sorted['salary_in_usd'].mean(),2)

ct_df_sorted = df[df['employment_type'] == 'CT'].sort_values(by='salary_in_usd', ascending=False)
ct_top = ct_df_sorted[:10]
ct_vid = round(ct_df_sorted['salary_in_usd'].mean(),2)

fl_df_sorted = df[df['employment_type'] == 'FL'].sort_values(by='salary_in_usd', ascending=False)
fl_top = fl_df_sorted[:10]
fl_vid = round(fl_df_sorted['salary_in_usd'].mean(),2)

pt_df_sorted = df[df['employment_type'] == 'PT'].sort_values(by='salary_in_usd', ascending=False)
pt_top = pt_df_sorted[:10]
pt_vid = round(pt_df_sorted['salary_in_usd'].mean(),2)

# print(pt_vid)
```

**2.3 Highest Paid Job Titles According to Different Employment Types**

```python
ft_vid_atl = ft_df_sorted.groupby('job_title').mean()['salary_in_usd'].round(2).sort_values(ascending = False)
top_ft_vid = ft_vid_atl[:5]
top_ft_vid
```

```python
top_ft_vid.plot(kind='barh', width = 0.8, color = "darkgreen")
plt.title('TOP 5 Job Titles among Full Time Employees with Highest Average Salaries in USD', fontdict={'fontweight':'bold', 'fontsize': 10})
plt.gca().invert_yaxis()
plt.ylabel(' ')
plt.show()
```

```python
ct_vid_atl = ct_df_sorted.groupby('job_title').mean()['salary_in_usd'].round(2).sort_values(ascending = False)
top_ct_vid = ct_vid_atl[:5]
top_ct_vid
```

```python
top_ct_vid.plot(kind='barh', width = 0.8, color = "darkgreen")
plt.title('TOP 5 Job Titles  among Contractors with Highest Average Salaries in USD', fontdict={'fontweight':'bold', 'fontsize': 10})
plt.gca().invert_yaxis()
plt.ylabel(' ')
plt.show()
```

```python
fl_vid_atl = fl_df_sorted.groupby('job_title').mean()['salary_in_usd'].round(2).sort_values(ascending = False)
top_fl_vid = fl_vid_atl[:5]
top_fl_vid
```

```python
top_fl_vid.plot(kind='barh', width = 0.8, color = "darkgreen")
plt.title('TOP 5 Job Titles  among Freelancers with Highest Average Salaries in USD', fontdict={'fontweight':'bold', 'fontsize': 10})
plt.gca().invert_yaxis()
plt.ylabel(' ')
plt.show()
```

```python
pt_vid_atl = pt_df_sorted.groupby('job_title').mean()['salary_in_usd'].round(2).sort_values(ascending = False)
top_pt_vid = pt_vid_atl[:5]
top_pt_vid
```

```python
top_pt_vid.plot(kind='barh', width = 0.8, color = "darkgreen")
plt.title('TOP 5 Job Titles  among Part Time Employees with Highest Average Salaries in USD', fontdict={'fontweight':'bold', 'fontsize': 10})
plt.gca().invert_yaxis()
plt.ylabel(' ')
plt.show()
```

```python
d_ft = dict(ft_vid_atl[:1])
rft = list(d_ft.keys())
rft[0] = rft[0] + ' (FT)'
vft = list(d_ft.values())

d_ct = dict(ct_vid_atl[:1])
rct = list(d_ct.keys())
rct[0] = rct[0] + ' (CT)'
vct = list(d_ct.values())

d_fl = dict(fl_vid_atl[:1])
rfl = list(d_fl.keys())
rfl[0] = rfl[0] + ' (FL)'
vfl = list(d_fl.values())

d_pt = dict(pt_vid_atl[:1])
rpt = list(d_pt.keys())
rpt[0] = rpt[0] + ' (PT)'
vpt = list(d_pt.values())

plt.bar(rft, vft, color = 'darkgreen')
plt.bar(rct, vct, color = 'lime')
plt.bar(rfl, vfl, color = 'olivedrab')
plt.bar(rpt, vpt, color = 'aquamarine')

plt.title('Comparison of Highest Paid Job Titles according to Different Employment Types ', fontdict={'fontweight':'bold', 'fontsize': 10})
plt.xticks(rotation=90)
plt.show()
```

**3.** **Variation of Salaries in Time Perpective**

```python
ft_vid_atl_metai = ft_df_sorted.groupby('work_year').mean()['salary_in_usd'].round(2)
ct_vid_atl_metai = ct_df_sorted.groupby('work_year').mean()['salary_in_usd'].round(2)
fl_vid_atl_metai = fl_df_sorted.groupby('work_year').mean()['salary_in_usd'].round(2)
pt_vid_atl_metai = pt_df_sorted.groupby('work_year').mean()['salary_in_usd'].round(2)

d_ft_m = dict(ft_vid_atl_metai)
rft_m = list(d_ft_m.keys())
rft_m[0] = str(rft_m[0]) + ' (FT)'
rft_m[1] = str(rft_m[1]) + ' (FT)'
rft_m[2] = str(rft_m[2]) + ' (FT)'
rft_m[3] = str(rft_m[3]) + ' (FT)'
vft_m = list(d_ft_m.values())

d_ct_m = dict(ct_vid_atl_metai)
rct_m = list(d_ct_m.keys())
rct_m[0] = str(rct_m[0]) + ' (CT)'
rct_m[1] = str(rct_m[1]) + ' (CT)'
rct_m[2] = str(rct_m[2]) + ' (CT)'
rct_m[3] = str(rct_m[3]) + ' (CT)'
vct_m = list(d_ct_m.values())

d_fl_m = dict(fl_vid_atl_metai)
rfl_m = list(d_fl_m.keys())
rfl_m[0] = str(rfl_m[0]) + ' (FL)'
rfl_m[1] = str(rfl_m[1]) + ' (FL)'
rfl_m[2] = str(rfl_m[2]) + ' (FL)'
rfl_m[3] = str(rfl_m[3]) + ' (FL)'
vfl_m = list(d_fl_m.values())

d_pt_m = dict(pt_vid_atl_metai)
rpt_m = list(d_pt_m.keys())
rpt_m[0] = str(rpt_m[0]) + ' (PT)'
rpt_m[1] = str(rpt_m[1]) + ' (PT)'
rpt_m[2] = str(rpt_m[2]) + ' (PT)'
rpt_m[3] = str(rpt_m[3]) + ' (PT)'
vpt_m = list(d_pt_m.values())
```

```python
plt.bar(rft_m, vft_m, color = 'darkgreen')
plt.bar(rct_m, vct_m, color = 'lime')
plt.bar(rfl_m, vfl_m, color = 'olivedrab')
plt.bar(rpt_m, vpt_m, color = 'aquamarine')

plt.title('Average Salaries according to Different Employment Types in 2020-2023 ', fontdict={'fontweight':'bold', 'fontsize': 10})
plt.xticks(rotation=90)

plt.show()
```

**4. Impact of Experience to the Salary**

```python
ft_vid_atl_exp = ft_df_sorted.groupby('experience_level').mean()['salary_in_usd'].round(2)
ct_vid_atl_exp = ct_df_sorted.groupby('experience_level').mean()['salary_in_usd'].round(2)
fl_vid_atl_exp = fl_df_sorted.groupby('experience_level').mean()['salary_in_usd'].round(2)
pt_vid_atl_exp = pt_df_sorted.groupby('experience_level').mean()['salary_in_usd'].round(2)

d_ft_exp = dict(ft_vid_atl_exp)
rft_exp = list(d_ft_exp.keys())
vft_exp = list(d_ft_exp.values())

d_ct_exp = dict(ct_vid_atl_exp)
rct_exp = list(d_ct_exp.keys())
vct_exp = list(d_ct_exp.values())

d_fl_exp = dict(fl_vid_atl_exp)
rfl_exp = list(d_fl_exp.keys())
vfl_exp = list(d_fl_exp.values())

d_pt_exp = dict(pt_vid_atl_exp)
rpt_exp = list(d_pt_exp.keys())
vpt_exp = list(d_pt_exp.values())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 6))

# plt.ylim(410000, 25000)  # Set the y-axis limits from 410,000 to 25,000

axs[0].bar(rct_exp, vct_exp, color='darkgreen')
axs[1].bar(rft_exp, vft_exp, color='lime')

axs[0].set_title('CT')
axs[1].set_title('FT')

for i in range(1, len(axs)):
    axs[i].set_ylim(axs[0].get_ylim())  # Align axes

fig.tight_layout()
plt.show()
```
